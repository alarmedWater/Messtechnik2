# 7.1 Auswertung Wiederholungskalibrierug
## 7.1.1 Auswertung analog elektrischer Druckaufnehmer

```python
p_elek = drucksensor - 0.00445
empfindlichkeit = 1/6 
p_elek = p_elek / empfindlichkeit
print(p_elek)
```

```python
plt.title("P_elek_analog ") 
plt.ylabel("Elektischer Druckaufnemer in [V]") 
plt.xlabel("Eingestellte Druck [bar]") 


plt.plot( drucksensor, p_elek, '-',label ='Messdaten')
plt.plot( drucksensor, p_elek, '*')


plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.title("Abeweichungsdiagramm")
plt.xlabel("Solldruck")
plt.ylabel("Abweichung")

diff_1 = p_elek - druckwaage
plt.bar(druckwaage, diff_1, color = 'r', width = 0.25, label = 'dist_ir')

plt.grid()
plt.legend()
plt.show()
```
